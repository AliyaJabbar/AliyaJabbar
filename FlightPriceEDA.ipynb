```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#matplotlib inline
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head(5)
```

```python
df.info()
```

```python
##feature Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
##Drop Date of journey
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head(5)
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]

```

```python
df.head(2)
```

```python
# Step 1: Clean 'Arrival_Time' to keep only the time part (ignore anything after space)
df['Arrival_Time_Clean'] = df['Arrival_Time'].str.split(' ').str[0]

# Step 2: Extract hour and minute
df['Arrival_hour'] = df['Arrival_Time_Clean'].str.split(':').str[0].astype(int)
df['Arrival_min'] = df['Arrival_Time_Clean'].str.split(':').str[1].astype(int)
```

```python
df.head(2)
```

```python
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.info()
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
```

```python
#df.drop('Dep_Time',axis=1,inplace=True)
```
